```python
import language_tool_python 
```

```python
import pandas as pd
```

```python
df = pd.read_csv("review_data.csv")
```

```python
df.head()
```

```python
my_tool = language_tool_python.LanguageTool('en-US') 
```

```python
d = []

for i in range(0, df.shape[0]):
    text = str(df.iloc[i][0])
    correct_text = str(my_tool.correct(text))
    if text == correct_text:
        d.append(1)
    else:
        d.append(0)
```

```python
df["new"] = d
```

```python
l = []

for i in range(0, df.shape[0]):
    if df.iloc[i][4] == 0:
        l.append("incorrect grammer")
    else:
        l.append("correct grammer")
```

```python
df["grammer_check"] = l
```
